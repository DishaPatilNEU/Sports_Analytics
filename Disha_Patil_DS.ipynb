```r
# In this section, you define two functions: 'adjust_start_games' and 'categorize_player'.
# The first function adjusts player statistics based on the number of games played, while the second function categorizes players based on their achievements.

# Define a function to adjust minutes and games started
adjust_start_games <- function(mins, games_start, games) {
  if (games < 82) {
    mins_adjusted <- mins * (82 / games)
    games_start_adjusted <- games_start * (82 / games)
  } else {
    mins_adjusted <- mins
    games_start_adjusted <- games_start
  }
  return(list(mins_adjusted = mins_adjusted, games_start_adjusted = games_start_adjusted))
}
# Here, the 'adjust_start_games' function is applied to calculate adjusted minutes and games started for each player in 'team_df'.
# The 'categorize_player' function is applied to categorize each player's season outcome, such as 'Elite' or 'Out of the League'.
# These transformations add new columns to 'team_df'.
# Define a function to categorize players
categorize_player <- function(All_NBA_First_Team, All_NBA_Second_Team, All_NBA_Third_Team,
                               Bill_Russell_NBA_Finals_MVP, all_star_game,
                               mins_adjusted, games_start_adjusted) {
  if (All_NBA_First_Team == 1 || All_NBA_Second_Team == 1 || All_NBA_Third_Team == 1 || Bill_Russell_NBA_Finals_MVP == 1) {
    return("Elite")
  } else if (all_star_game == TRUE) {
    return("All-Star")
  } else if (games_start_adjusted >= 41 || mins_adjusted >= 2000) {
    return("Starter")
  } else if (mins_adjusted >= 1000) {
    return("Rotation")
  } else if (mins_adjusted > 0) {
    return("Roster")
  } else {
    return("Out of the League")
  }
}


# Apply the adjust_start_games function and create new columns for adjusted minutes and games started
team_df <- team_df %>%
  rowwise() %>%
  mutate(AdjustedMinutes = adjust_start_games(mins, games_start, games)[[1]],
         AdjustedGamesStart = adjust_start_games(mins, games_start, games)[[2]])

team_df$All.NBA.First.Team[is.na(team_df$All.NBA.First.Team)] <- 0
team_df$All.NBA.Second.Team[is.na(team_df$All.NBA.Second.Team)] <- 0
team_df$All.NBA.Third.Team[is.na(team_df$All.NBA.Third.Team)] <- 0
team_df$Bill.Russell.NBA.Finals.MVP[is.na(team_df$Bill.Russell.NBA.Finals.MVP)] <- 0
team_df$all_star_game[is.na(team_df$all_star_game)] <- FALSE

# Apply the categorization function to each row and create a new column 'Season Outcome'
team_df <- team_df %>%
  rowwise() %>%
  mutate(SeasonOutcome = categorize_player(All.NBA.First.Team, All.NBA.Second.Team, All.NBA.Third.Team,
                                          Bill.Russell.NBA.Finals.MVP, all_star_game,
                                          AdjustedMinutes, AdjustedGamesStart))

# View the resulting data frame
View(team_df)
library(base)
Visualizations_df<-  data.frame(team_df)
```

```r

# -----------------

  # Modify the calculate_highest_career_outcome function
# This function calculates the highest career outcome based on a set of outcomes and their priorities.
calculate_highest_career_outcome <- function(outcomes) {
  # Define the priority order for outcomes
  outcome_order <- c('Elite', 'All-Star', 'Starter', 'Rotation', 'Roster', 'Out of the League')
  # Create a data frame to count the occurrences of each outcome
  outcome_count <- data.frame(Outcome = outcome_order, Count = 0)

  # Check if outcomes is empty or contains invalid values
  if (length(outcomes) == 0 || !all(outcomes %in% outcome_order)) {
    return('Out of the League')
  }

  for (outcome in outcomes) {
    # Increment the count for the current outcome
    outcome_count$Count[outcome_count$Outcome == outcome] <- outcome_count$Count[outcome_count$Outcome == outcome] + 1
    # Adjust counts based on priority levels
    if (outcome == 'Elite') {
      outcome_count$Count[outcome_count$Outcome %in% c('All-Star', 'Starter', 'Rotation', 'Roster', 'Out of the League')] <- outcome_count$Count[outcome_count$Outcome %in% c('All-Star', 'Starter', 'Rotation', 'Roster', 'Out of the League')] + 1
    } else if (outcome == 'All-Star') {
      outcome_count$Count[outcome_count$Outcome %in% c('Starter', 'Rotation', 'Roster', 'Out of the League')] <- outcome_count$Count[outcome_count$Outcome %in% c('Starter', 'Rotation', 'Roster', 'Out of the League')] + 1
    } else if (outcome == 'Starter') {
      outcome_count$Count[outcome_count$Outcome %in% c('Rotation', 'Roster', 'Out of the League')] <- outcome_count$Count[outcome_count$Outcome %in% c('Rotation', 'Roster', 'Out of the League')] + 1
    } else if (outcome == 'Rotation') {
      outcome_count$Count[outcome_count$Outcome %in% c('Roster', 'Out of the League')] <- outcome_count$Count[outcome_count$Outcome %in% c('Roster', 'Out of the League')] + 1
    } else if (outcome == 'Roster') {
      outcome_count$Count[outcome_count$Outcome == 'Out of the League'] <- outcome_count$Count[outcome_count$Outcome == 'Out of the League'] + 1
    }
  }
  # Determine the highest career outcome based on counts and priorities
  highest_outcome <- 'Out of the League'
  for (outcome in outcome_order) {
    if (outcome_count$Count[outcome_count$Outcome == outcome] >= 2) {
      highest_outcome <- outcome
      break
    }
  }

  return(highest_outcome)
}



# ----

# Define a function to get the highest priority outcome
get_highest_priority_outcome <- function(group) {
  priority_levels <- c('Elite' = 6, 'All-Star' = 5, 'Starter' = 4, 'Rotation' = 3, 'Roster' = 2, 'Out of the League' = 1)

  max_priority <- 0
  max_outcome <- NULL

  for (outcome in group$SeasonOutcome) {
    priority <- priority_levels[outcome]
    if (priority > max_priority) {
      max_priority <- priority
      max_outcome <- outcome
    }
  }

  return(max_outcome)
}
# This section groups players by ID and calculates their highest career outcome using the defined functions.
# It loops through unique player IDs, processes their data, and stores the highest career outcome in 'career_outcomes' data frame.
# Group and calculate the highest career outcome
unique_players <- unique(team_df$nbapersonid)
career_outcomes <- data.frame()


for (player_id in unique_players) {
  player_data <- team_df %>% filter(nbapersonid == player_id) %>% select(nbapersonid, SeasonOutcome, season)
  player_data <- player_data %>% group_by(nbapersonid, season) %>% do(Highest_Season_Outcome = get_highest_priority_outcome(.))
  player_data <- player_data %>% distinct(nbapersonid, season, .keep_all = TRUE)

  player_data <- player_data %>% arrange(season, .by_group = TRUE)
  player_data <- player_data[5:nrow(player_data), ]  # Exclude the first 4 rows

  highest_outcome <- calculate_highest_career_outcome(player_data$Highest_Season_Outcome)

  career_outcomes <- rbind(career_outcomes, data.frame(highest_career_outcome = highest_outcome, nbapersonid = player_id))
}

# Print the resulting data frame
career_outcomes
```

In this section, we want to predict a team's offensive rebounding percentage in their next game using the rebounding_data dataset. To make this prediction, we'll calculate the average offensive rebounding percentage from all the previous games. Offensive rebounding percentage at a single game level is calculated as the number of offensive rebounds divided by the number of offensive rebound chances, where offensive rebound chances represent the team's missed shots. When considering multiple games, we'll calculate the total number of offensive rebounds and divide it by the total number of offensive rebound chances.

Now, let's determine the predicted offensive rebound percentage for the Oklahoma City Thunder (OKC) in game 81 using the data from games 1 to 80.
